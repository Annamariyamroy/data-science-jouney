UNIT VECTOR  
 A unit vector is the same arrow **shrunk to length exactly 1** — same direction, but magnitude = 1.

Why is this useful? Because sometimes you only care about **which direction** something points, not how far. For example in recommendation systems — you care about whether two users like *similar things*, not how many ratings they gave.

To get a unit vector you just divide by the magnitude:

unit = u / |u| = [3/5, 4/5] = [0.6, 0.8]




````python
import numpy as np
u= np.array([3.0,4.0])
print(f"vector u: {u}")
print(f"magnitude |u| : {np.linalg.norm(u)}")
u_unit= u/np.linalg.norm(u)
print(f"unit vector: {u_unit}")
print(f"unit_vector_magnitude: {np.linalg.norm(u_unit)}")
#np.allclose(a,b)checks whether bpth are almost equal.returns True or False
np.allclose(np.linalg.norm(u_unit),1)

#------------------------------------------------
v = np.array([1.0, 2.0, 3.0])
v_unit= v/np.linalg.norm(v)
print(np.linalg.norm(v_unit)==1)
np.allclose(np.linalg.norm(v_unit),1)
````

COSINE SIMILARITY
Cosine similarity between two vectors.
Returns value between -1 (opposite) and 1 (identical direction).
Cosine similarity measures how similar two vectors are based on their direction, not their magnitude.

Formula:cosine similarity=a⋅b/|a||b|
It is mathematically the same as:cos(θ) , where θ is the angle between two vectors.
Range of values

+1 → vectors point in same direction (very similar)
0 → vectors are perpendicular (no similarity)
−1 → vectors point in opposite directions (completely different)

->Used in NLP, recommendation systems, clustering
->If one vector = constant × another → cosine similarity = 1
eg.[1,1]=a
[10,10]=b
->If one vector = negative multiple of another → cosine similarity = −1
->If dot product is 0 ->cosine similarity=0



````python
def cosine_similarity(a,b):
    return(a @ b/(np.linalg.norm(a)*np.linalg.norm(b)))
doc1 = np.array([3, 2, 1, 0, 0])   # data science article
doc2 = np.array([2, 3, 2, 0, 0])   # another data science article
doc3 = np.array([0, 0, 0, 3, 2])   # cooking article

print("doc1 vs doc2 (both data science):")
print(f"Cosine similarity: {cosine_similarity(doc1,doc2):.4f} high similarity ")
print("doc1 vs doc3 (completely different topics):")
print(f"Cosine similarity: {cosine_similarity(doc1,doc3)} no similarity")
print("doc1 vs doc1 (identical):")
print(f"Cosine similarity: {cosine_similarity(doc1,doc1)} identical")


````

Projection
-> Projection of vector a onto vector b means:
-> How much of vector a lies along the direction of vector b
->It is like the shadow of vector a along direction b.
Projection formula
proj of a onto b = (a·b / b·b) × b
   a·b      → how much a points in b's direction (dot product)
   b·b      → magnitude of b squared (|b|²)
   a·b/b·b  → a scalar — how many "units of b" does a cover
    × b      → scale b by that scalar to get the actual vector
Example

Project:

a = [4,3]
b = [1,0]
proj = [4,0] This keeps only the x-direction component of vector a.

Perpendicular component
A vector can be split into:

a = projection + perpendicular component

Formula:

perpendicular = a − projection

Example:

a = [4,3]
projection = [4,0]
perpendicular = [0,3] ,the perpendicular component represents the part of a not aligned with b.

Projection and perpendicular component are always orthogonal (perpendicular):

projection · perpendicular = 0

````python
def project(a,b):
    scalar= (a @ b)/(b @ b)
    return scalar * b

a = np.array([4.0, 3.0])
b = np.array([1.0, 0.0])   # x-axis
projection=project(a,b)
print(f"Projection of {a} on {b} is {projection}")

perp= a-projection
print(f"The perpendicular component is {perp}")
print(f"Are they perpendicular? {abs(projection @ perp) < 1e-10}")#checking if dot product is zero
````

Eigenvalues and eigenvectors 
#For a matrix A, an eigenvector v satisfies: A @ v = λ × v
#where λ is the eigenvalue (a scalar)
#Translation: multiplying v by A only SCALES it, doesn't rotate it
#The eigenvector points in a "special" direction that A preserves

````python
A = np.array([[4., 1.],
              [2., 3.]])
eigenvalues,eigenvectors= np.linalg.eig(A)
print("eigen values:", eigenvalues)
print("eigen vectors:",eigenvectors)

# Each COLUMN is one eigenvector
# eigenvectors[:,0] corresponds to eigenvalues[0]
# eigenvectors[:,1] corresponds to eigenvalues[1]

#verification
for i in range (len(eigenvalues)):
    v= eigenvectors[:,i]
    lam= eigenvalues[i]
    Av= A @ v
    lam_v= lam * v
    
    print(f"\nEigenvector {i}: {v.round(2)}, lambda: {lam:.1f}")
    print(f"A @ v : {Av.round(3)}")
    print(f"lam_v: {lam_v.round(3)}")
    print(f"verifiying eigen vector : {np.allclose(Av,lam_v)}")
    
````

PRINCIPAL COMPONENT ANALYSIS

What is an eigenvalue
An eigenvalue tells you how much a matrix stretches its eigenvector in a special direction without rotating it. In data science, the eigenvalues of a covariance matrix tell you how much variance exists in each principal direction — a large eigenvalue means that direction contains a lot of information, a small eigenvalue means that direction is mostly noise.

The eigenvalue tells you how much the matrix STRETCHES the eigenvector
In PCA: eigenvectors of covariance matrix = principal components
eigenvalues = how much variance each component captures
Large eigenvalue → that direction captures a lot of variance in data
Small eigenvalue → that direction is mostly noise
Covariance matrix size is: (number of features)×(number of features)
x1
x2
[cov(x1,x1)cov(x2,x1)
cov(x1,x2)cov(x2,x2)]


Key Applications of PCA:

Dimensionality Reduction
Reduces number of features while keeping maximum important information.
Helps simplify large datasets before applying machine learning models.
Data Visualization
Converts high-dimensional data (e.g., 10D, 100D) into 2D or 3D for plotting and understanding patterns.
Noise Reduction
Removes components with very low variance which mostly represent noise in data.
Feature Extraction
Creates new meaningful features (principal components) from existing correlated features.
Speed Improvement in Machine Learning
Reduces computation time by decreasing number of input variables.
Avoiding Overfitting
Eliminates redundant features and improves model generalization.
Applications in Real-world Domains
Face recognition (Eigenfaces)
Image compression
Signal processing
Bioinformatics (gene data analysis)
Finance (risk analysis)
Natural Language Processing embeddings preprocessing

Summary:
PCA helps transform complex high-dimensional datasets into simpler representations while preserving most of the useful information, making machine learning models faster, cleaner, and more efficient.

````python
import numpy as np
rng=np.random.default_rng(seed=42)
n=200
x1=rng.normal(0,3,n)
x2=0.9 * x1 + rng.normal(0, 0.8, n) # strongly correlated with x1
data= np.column_stack((x1,x2))#shape(200,2)

# Step 1: Centre the data
data_centered=data-np.mean(data,axis=0)#(200,2)

# Step 2: Covariance matrix
# Note: np.cov expects (n_features, n_samples) so transpose
#shape will always be(n_features)
cov=np.cov(data_centered.T)
print(f"\n covarience matrix : {cov.round(3)}")

# Step 3: Eigendecomposition
# Use eigh (not eig) for symmetric matrices — more stable
#co varince matrices are always symmetric
evals,evecs= np.linalg.eigh(cov)
print(f"\nEigen values(unsorted): {evals.round(3)}")
print(f"Eigen vectors :{evecs.round(3)}")

# Step 4: Sort by eigenvalue descending
order= np.argsort(evals)[::-1]
evals=evals[order]
evecs= evecs[:,order]
print(f"\n eigen values sorted: {evals}")

#variance explained
total_variance=evals.sum()
variance_explained= (evals/total_variance)*100
for i,(ev,pct) in enumerate(zip(evals,variance_explained)):
    print(f"Component {i+1}: {ev:.4f} {pct:.1f}%")
# Component 1 should explain ~95%+ — the dominant direction

# Step 5: Project data onto first principal component
projected= data_centered @ evecs#(200,2)@(2,2)
print(f"\nOriginal shape:  {data.shape}")
print(f"Projected shape: {projected.shape}")

# Using only PC1 gives you 1D representation preserving most variance
print(f"\nVariance in PC1: {projected[:,0].round(4)}")
print(f"variance in PC2: {projected[:,1].round(4)}")
pc1_only= projected[:,0]
print(pc1_only)
#Now dataset becomes:

'''2D → 1D

But still keeps:

~97% information

This is dimensionality reduction.'''
````

LINEAR REGRESSION - from scratch using the normal equation
Linear Regression: Slope and Intercept

The basic equation of linear regression is:

$$
y = mx + b
$$

Where:

- **m (slope)** → rate at which \( y \) changes when \( x \) changes  
- **b (intercept)** → value of \( y \) when \( x = 0 \)

In Machine Learning notation:

$$
y = w_1 x + w_0
$$

Where:

- \( w_1 \) = slope (coefficient)
- \( w_0 \) = intercept (bias)

---

 What is Slope?

Slope represents how much the output changes when the input increases by **1 unit**.

Example:

$$
y = 3x + 5
$$

Slope = **3**

Meaning:

If \( x \) increases by 1 → \( y \) increases by 3

Example values:

| x | y |
|---|---|
| 1 | 8 |
| 2 | 11 |
| 3 | 14 |

Each time \( x \) increases by 1, \( y \) increases by 3.

---

What is Intercept?

Intercept represents the value of \( y \) when:

$$
x = 0
$$

Example:

$$
y = 3x + 5
$$

When \( x = 0 \):

$$
y = 5
$$

So the intercept is **5**.

This is the point where the regression line crosses the **y-axis**.

---

 Visual Intuition

Example:

$$
y = 2x + 4
$$

- slope = **2**
- intercept = **4**

Meaning:

- Graph starts at \( y = 4 \) when \( x = 0 \)
- Then increases by **2 units** for every increase of **1 unit** in \( x \)

Linear Regression — Normal Equation (Short Note)
Input (X) = feature matrix
Output (y) = target values
Model assumes:
y=w0+w1x
Model learns weights (w) = intercept + slope(s)
Computed using normal equation:Since perfect solution rarely exists (because of noise), it finds best approximation using,This minimizes prediction error.:
w=(XTX)−1XTy
Shapes rule:
X shape = (m, n)
X.T @ X = (n, n)   → square matrix
X.T @ y = (n,)
w shape = (n,)
Meaning:
X  → inputs
y  → correct outputs
w  → learned relationship between them
Xw → predictions


### Model Evaluation Metrics in Linear Regression

After making predictions using:

y_pred = X @ w

we evaluate how well the model performs using the following metrics:

---

### 1. Residuals

Residuals measure the difference between actual and predicted values.

Residual = actual value − predicted value

residuals = y − y_pred

Each residual represents the prediction error for one data point.

---

### 2. MSE (Mean Squared Error)

Formula:

MSE = mean((y − y_pred)²)

Code:

mse = np.mean(residuals**2)

Meaning:

Average squared prediction error.  
Penalizes large errors more strongly.  
Lower MSE = better model.

---

### 3. RMSE (Root Mean Squared Error)

Formula:

RMSE = √MSE

Code:

rmse = np.sqrt(mse)

Meaning:

Average prediction error in the same units as the target variable.  
Easy to interpret.

---

### 4. MAE (Mean Absolute Error)

Formula:

MAE = mean(|y − y_pred|)

Code:

mae = np.mean(np.abs(residuals))

Meaning:

Average absolute prediction error.  
Treats all errors equally.

---

### 5. R² Score (Coefficient of Determination)

Formula:

R² = 1 − (SS_residual / SS_total)

Code:

ss_residual = np.sum(residuals**2)
ss_total = np.sum((y − np.mean(y))**2)
r_squared = 1 − (ss_residual / ss_total)

Meaning:

Measures how well the model explains variance in the data.

R² range:

1 → perfect prediction  
0 → model no better than predicting mean  
< 0 → poor model performance

Higher R² = better model

````python
# Step 1: Build the design matrix X
# Add a column of 1s for the intercept term (bias)
# Each row is [1, x_i] — the 1 gets multiplied by the bias weight
# True relationship: y = 3x + 5 + noise
# We want our model to learn the slope (3) and intercept (5)
rng=np.random.default_rng(seed=42)
n=100
x= rng.uniform(1,10,n) #shape (100,)
y = 3 * x + 5 + rng.normal(0, 2, n) #shape y=(100,)

X=np.column_stack((np.ones(n),x))
print(f"Design matrix shape : {X.shape}")
print(f"First five rows \n:{X[:5,:]}")

# Step 2: Apply the normal equation
# w = (X.T @ X)^{-1} @ X.T @ y

XtX= X.T @ X
XtX_inv= np.linalg.inv(XtX)#(2,100)
Xty= X.T @ y # shape(2,) #[sum(y)sum(x*y)]
w= XtX_inv @ Xty
print("Learned weights:")
print(f"Intercept: {w[0]:.4f} (true:5.0)")
print(f"slope: {w[1]:.4f} (true:3.0)")

# Step 3: Make predictions
y_pred= X @ w #shape (100,2) @ (2) = (100,) 
print(f"First 5 prediction:\n {y_pred[:5]}")
print(f"First 5 actual : \n{y[:5]}")

# Step 4: Evaluate the model
residuals= y-y_pred

# Mean Squared Error
mse= np.mean(residuals**2)
rmse=np.sqrt(mse)

# Mean Absolute Error
mae= np.mean(np.abs(residuals))

# R² score — proportion of variance explained by the model
# R² = 1 means perfect fit, R² = 0 means model is no better than predicting mean
ss_residual = np.sum(residuals**2)          # variance unexplained
ss_total    = np.sum((y - np.mean(y))**2)
r_squared   = 1 - (ss_residual / ss_total)

print(f"\nModel Evaluation:")
print(f"  MSE:   {mse:.4f}")
print(f"  RMSE:  {rmse:.4f}")
print(f"  MAE:   {mae:.4f}")
print(f"  R²:    {r_squared:.4f}")
````

````python
# Step 5: Extend to multiple features
# Now y depends on multiple x variables — same formula, bigger X
rng=np.random.default_rng(seed=42)
n=200
x1= rng.normal(0,5,n)
x2= rng.normal(0,3,n)
x3=rng.normal(0,8,n)

y_multi = 2*x1 + (-3)*x2 + 4*x3 + 7 + rng.normal(0, 2, n)

x_multi= np.column_stack((np.ones(n),x1,x2,x3))#(200,4)
w_multi=np.linalg.inv((x_multi.T @ x_multi))@ (x_multi.T @ y_multi) #w=(XTX)−1XTy

print("Learned coefficients:\n")
print(f"Intercept: {w_multi[0]:.4f} true:7")
print(f"x1 coefficient: {w_multi[1]:.4f} true : 2")
print(f"x2 coefficient: {w_multi[2]:.4f} true : -3")
print(f"x3 coefficient: {w_multi[3]:.4f} true : 4")


````

np.linalg.lstsq() — Least Squares Solver (Linear Regression)

Function:

w_lstsq, residuals_sq, rank, sv = np.linalg.lstsq(X, y, rcond=None)

This function solves the equation:

Xw ≈ y

and returns the best-fit regression coefficients using the **least squares method**.

It returns 4 outputs:

1️⃣ w_lstsq (Weights / Coefficients)

Best-fit regression parameters.

Example:
[intercept, x1_weight, x2_weight, x3_weight]

Equivalent to:

w = (XᵀX)⁻¹ Xᵀy

but more stable and safer.

---

2️⃣ residuals_sq (Sum of Squared Residuals)

Value of:

sum((y − y_pred)²)

Measures total prediction error remaining after fitting the model.

Lower value → better fit

May return empty array if solution is exact.

---

3️⃣ rank (Rank of Matrix X)

Number of independent columns (features) in matrix X.

If:

rank = number of columns in X

→ features are linearly independent  
→ matrix is full rank  
→ solution is stable and unique

---

4️⃣ sv (Singular Values of X)

Singular values from SVD decomposition of X.

Used to check:

- numerical stability
- feature redundancy
- multicollinearity

Very small singular values → possible feature dependence
