A **vector** is a list of numbers that represents a **direction and magnitude in space** — 
like an arrow drawn from the origin:
y
↑
4 |     * ← point (3, 4)
  |    /
  |   /  ← this arrow IS the vector [3, 4]
  |  /
  | /
  |/________→ x
  0    3
  
  In NumPy a vector is just a 1D array
  Length of the arrow is called magnitude or norm of the vector
  In NumPy it is np.linalg.norm()
  
UNIT VECTOR  
 A unit vector is the same arrow **shrunk to length exactly 1** — same direction, but magnitude = 1.

Why is this useful? Because sometimes you only care about **which direction** something points, not how far. For example in recommendation systems — you care about whether two users like *similar things*, not how many ratings they gave.

To get a unit vector you just divide by the magnitude:

unit = u / |u| = [3/5, 4/5] = [0.6, 0.8]




```python
import numpy as np
u= np.array([3.0,4.0])
print(f"vector u: {u}")
print(f"magnitude |u| : {np.linalg.norm(u)}")
u_unit= u/np.linalg.norm(u)
print(f"unit vector: {u_unit}")
print(f"unit_vector_magnitude: {np.linalg.norm(u_unit)}")
#np.allclose(a,b)checks whether bpth are almost equal.returns True or False
np.allclose(np.linalg.norm(u_unit),1)

#------------------------------------------------
v = np.array([1.0, 2.0, 3.0])
v_unit= v/np.linalg.norm(v)
print(np.linalg.norm(v_unit)==1)
np.allclose(np.linalg.norm(v_unit),1)
```

VECTOR NORMS IN DIFFERENT WAYS
L1 Norm (Manhattan distance) : l1_norm = np.linalg.norm(v, ord=1)
-> # sum of absolute values: 3+4+2=9
-> Used in Lasso Regression
L2 Norm (Euclidean distance): l2_norm = np.linalg.norm(v, ord=2)
-># sqrt(9+16+4) ≈ 5.385
-> actual straight-line distance from origin to vector.
-> Used in Ridge Regression
L∞ Norm (Maximum norm)
-> # max absolute value: 4
-> Only cares about the largest movement in any direction
-> Used in: optimisation problems
 

```python
v = np.array([3.0, -4.0, 2.0])
l1_norm=np.linalg.norm(v,ord=1)
l2_norm = np.linalg.norm(v, ord=2)
l3_norm=np.linalg.norm(v,ord=np.inf)
print(f"L1 norm: {l1_norm}")
print(f"L2 norm: {l2_norm}")
print(f"L3 norm: {l3_norm}")
```

```python
# Three equivalent ways to compute dot product
dot1 = np.dot(u, v)         # function form
dot2 = u @ v                # operator form — USE THIS in practice
dot3 = np.sum(u * v)        # manual: element-wise multiply then sum

import numpy as np
a= np.array([1.0, 0.0])
b = np.array([0.0, 1.0])
c = np.array([-1.0, 0.0])
print(a @ b)#0.0 perpendicular
print(a @ a)#1.0 similar direction
print(a @ c ) #-1.0 opposite direction
```

```python
u = np.array([3.0, 4.0])
v = np.array([1.0, 0.0])
w = np.array([-3.0, -4.0]) 

#Return angle between two vectors in degrees.
#clips the valule range(-1,1)
#np.arccos finds inverse of cosine and returns result in radians and n0t degrees
#np.degree convers radians to degree
def  angle_between(a,b):
    cos_theta= a @ b/np.linalg.norm(a) *np.linalg.norm(b)
    cos_theta= np.clip(cos_theta,-1,1)
    return np.degrees(np.arccos(cos_theta))
print(f"angle between u and v: {angle_between(u,v):.2f}") 
print(f"angle between u and u: {angle_between(u,u):.2f}")
print(f"angle between u and w: {angle_between(u,w):.2f}")
Angle	Meaning
0°   = identical direction
<90° = somewhat similar
90°  = unrelated
>90° = opposite direction
180° = completely opposite
```
