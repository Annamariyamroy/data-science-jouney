**MODE**
Mode is the most frequently occurring value. 
NumPy has no built in mode function — you implement it manually using np.unique.

arr = [3, 1, 2, 3, 4, 2, 3]

vals, counts = np.unique(arr, return_counts=True)

vals   = [1, 2, 3, 4]   ← unique values sorted
counts = [1, 2, 3, 1]   ← how many times each appears
mode = value with highest count
counts.argmax() = index 2
vals[2] = 3 → mode is 3

**Simple rule**
continuous data  (67.3, 67.4, 67.5...)  → mode is useless
discrete data    (1, 2, 3, 4, 5...)     → mode can be useful
categorical data (A, B, C / yes, no...) → mode is very useful

**Multimodal Distribution:** if multiple values share the highest count → all are modes

````python
rng= np.random.default_rng(seed=42)
integer_scores= rng.integers(50,101,size=50)
vals,counts= np.unique(integer_scores,return_counts=True)
mode_value=vals[np.argmax(counts)]
mode_count= np.max(counts)
print(f"mode:{mode_value} (appears {mode_count} times)")


all_modes = vals[counts == mode_count]
print(f"All modes: {all_modes}")
````

## Variance and Standard Deviation

### What they measure
How spread out values are from the mean.
- Low std → values clustered close to mean
- High std → values scattered far from mean

### Formulas
```python
# step by step manually:
deviations = data - np.mean(data)   # distance from mean
sq_dev     = deviations**2          # square to remove negatives
variance   = np.sum(sq_dev) / (n-1) # average squared deviation
std        = np.sqrt(variance)       # back to original units
```

### Population vs Sample
| | Formula | When to use |
|---|---|---|
| Population | divide by n | you have ALL data |
| Sample | divide by n-1 (ddof=1) | you have a sample |

**Rule: almost always use ddof=1 in practice**

Bessel's correction (n-1) compensates for sample mean being
slightly off from true population mean.

### Coefficient of Variation (CV)
Compares spread across datasets with different scales.
```python
cv = (std / mean) * 100
```
- CV < 15%  → low variability
- CV 15-30% → moderate variability  
- CV > 30%  → high variability

**Use case:** comparing spread of exam scores (0-100)
vs salaries (0-100000) on equal footing.

````python
var_pop= np.var(data)
var_sample= np.var(data,ddof=1)
std_pop= np.std(data)
std_sample=np.std(data,ddof=1)
print(f"population variance:{var_pop:.4f}")
print(f"sample variance:{var_sample:.4f}")
print(f"population standard:{std_pop:.4f}")
print(f"sample standard: {std_sample:.4f}")

#manual variance

'''population variance → divide by n
  use when: you have ALL the data (every student in the world)

sample variance → divide by n-1
  use when: you have a SAMPLE (200 students out of millions)'''
        
deviation= data-np.mean(data)
sq_dev = deviation**2
manual_variance= np.sum(sq_dev)/(len(sq_dev)-1)
print(f"manual variance:{manual_variance:.4f}")
print(f"isclose: {np.isclose(manual_variance,var_sample)}")

#Coefficient of variation

cv= (std_sample/mean)*100
print(f"Coefficient of variation:{cv:.2f}%")
````

**Skewness and Kurtosis**
symmetric (normal):          right skewed:
     *                            *
    ***                          ***
   *****                        *****
  *******                      *******
 *********                    ********.....
55   68   81                 55  68  81  150
     ↑                            ↑        ↑
   center                       center   long right tail
                                          pulling mean right
Skewness
skewness = 0    → perfectly symmetric
skewness > 0    → right tail (positive skew)
skewness < 0    → left tail (negative skew)

|skewness| < 0.5  → approximately symmetric
|skewness| > 1.0  → significantly skewed

Formula for Skewness
**skewness = mean of ((data - mean) / std)³** 

Squaring removes negatives (used in variance). Cubing keeps the sign — so left deviations give negative values, right deviations give positive values. The average tells you which side dominates.

**Kurtosis — measuring tail heaviness**
->how fat or thin the tails are compared to a normal distribution.

Formula 

**kurtosis = 0   → same tails as normal (mesokurtic)
kurtosis > 0   → fatter tails → more outliers (leptokurtic)
kurtosis < 0   → thinner tails → fewer outliers (platykurtic)**

Value type	Interpretation
Raw kurtosis > 3	heavier tails
Raw kurtosis = 3	normal tails
Raw kurtosis < 3	lighter tails
Excess kurtosis > 0	heavier tails
Excess kurtosis = 0	normal tails
Excess kurtosis < 0	lighter tails

**Demonstration with three distributions**

1. right skewed  → rng.exponential()
-> classic example: income, house prices, waiting times
-> most values small, few very large values

2. left skewed   → 100 - rng.exponential()
-> classic example: exam scores near max
-> most students score high, few score very low

3. your data     → rng.normal()
->approximately symmetric
->skewness close to 0

````python
deviations= data-np.mean(data)
std= np.std(data,ddof=1)

skewness= np.mean((deviations/std)**3)
kurtosis= np.mean((deviations/std)**4)

print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurtosis:.4f}")

right_skewed= rng.exponential(scale=5,size=500)
left_skewed= 100-rng.exponential(scale=5,size=500)

for name,dist in [("Right skewed",right_skewed),
                   ("Left skewed",left_skewed),
                   ("your data",data)]:
    dev= dist-np.mean(dist)
    std=np.std(dist)
    sk= np.mean((dev/std)**3)
    ku=np.mean((dev/std)**4)
    if sk>0.1:
        direction= "->right_tail"
    elif sk<-0.1:
        direction="->left_tail"
    else:
        direction= symmetric
    print(f"\n{name}")
    print(f"Skewness: {sk:.4f} {direction} ")
    print(f"kurtosis: {ku:.4f}")

````

````python
#Method 1 - Z score method
data_o= np.append(data, [150, 160, -10, 155])
mean_o= np.mean(data_o)
std_o=np.std(data_o)

z_scores= (data_o-mean_o)/std
outlier_mask_z=np.abs(z_scores)>3
print("\nZ Score (threshold=3)")
print(f"No of outliers: {np.sum(outlier_mask_z)}")
print(f"outlier values:{data_o[outlier_mask_z]}")
print(f"z_scores of outliers: {z_scores[outlier_mask_z]}")

#Method2 IQR method

q1_o= np.percentile(data_o,25)
q3_o=np.percentile(data_o,75)
iqr_o= q3_o-q1_o

#mild outliers
lower_fence= q1_o-(1.5*iqr_o)
upper_fence= q3_o+(1.5*iqr_o)

#extreme outliers
extreme_lower= q1_o-(3*iqr_o)
extreme_upper= q3_o+(3*iqr_o)

print("\nIQR METHOD")
print(f"Q1:{q1_o:.2f} Q3: {q3_o:.2f} IQR: {iqr_o:.2f}")
print(f"lower fence: {lower_fence:.2f}")
print(f"Upper fence: {upper_fence:.2f}")
print(f"Extreme lower:{extreme_lower:.2f}")
print(f"Extreme upper:{extreme_upper:.2f}")

mild_outliers= data_o[(data_o<lower_fence)|(data_o>upper_fence)]
print(f"length of mild outliers:{len(mild_outliers)} outlier_values:{mild_outliers}")
extreme_outliers=data_o[(data_o<extreme_lower)|(data_o>extreme_upper)]
print(f"length of extreme outliers: {len(extreme_outliers)} outlier_values: {extreme_outliers}")

#Method 3 Modified Z Score
def modified_z_score(arr):
    median_o= np.median(arr)
    mad= np.median(np.abs(arr-median_o))
    modified_z= 0.6745*((arr-median_o)/mad)
    return modified_z
mod_z=modified_z_score(data_o)
outlier_mask_modz= np.abs(mod_z)>3.5 
outliers= data_o[outlier_mask_modz]
print("\nModified Z Score Method")
print("Modified z score (threshold=3.5)")
print(f"No of outliers: {len(outliers)}")
print(f"outlier values: {outliers}")

# ── Compare all three methods ────────────────────────────────
print("\nMethod Comparison:")
print(f"{'Method':<25} {'Outliers Found':>15} {'Values'}")
print("─" * 65)

z_out   = data_o[np.abs(z_scores) > 3]
iqr_out = data_o[
    (data_o < lower_fence) | (data_o > upper_fence)
]
mz_out  = data_o[np.abs(mod_z) > 3.5]

for name, out in [("Z-score (>3)", z_out),
                  ("IQR (1.5×IQR)", iqr_out),
                  ("Modified Z (>3.5)", mz_out)]:
    vals = sorted(out.tolist())
    print(f"{name:<25} {len(out):>15}      {vals}")
````

**Correlation, Covariance & Histogram Analysis**

**Part 1-Covariance and correlation**
-> Covariance measures how two variables move together.

study_hours and exam_score:
  student studies 8 hrs → scores 85
  student studies 2 hrs → scores 40
  when study goes up, score goes up → POSITIVE covariance

study_hours and sleep_hours:
  student studies 8 hrs → sleeps 5 hrs
  student studies 2 hrs → sleeps 8 hrs
  when study goes up, sleep goes down → NEGATIVE covariance

study_hours and random_var:
  no relationship whatsoever → NEAR ZERO covariance

**Formula- 
cov(X, Y) = sum((xi - x_mean) × (yi - y_mean)) / (n-1)
**

(xi - x_mean) → how far each x value is from x mean
(yi - y_mean) → how far each y value is from y mean

multiply them together:
  both above mean → positive × positive = positive
  both below mean → negative × negative = positive
  one above, one below → positive × negative = negative

sum all products → tells you which direction dominates
divide by n-1   → sample covariance

**Pearson Correlation — fixes the scale problem**

Problem with covariance — it depends on scale
study hours range: 1-10
exam score range:  0-100

cov(study, exam) = 40.5   ← large number

study hours range: 1-10
sleep hours range: 3-10

cov(study, sleep) = -0.8  ← small number
**r = cov(X, Y) / (std(X) × std(Y))**

->np.corrcoef is not a separate method — it is just NumPy's built-in function that computes Pearson correlation for you.

->Dividing by both standard deviations removes the units. Result is always between -1 and +1:

r = +1.0  → perfect positive  (study doubles, score always doubles)
r = +0.8  → very strong positive
r = +0.5  → moderate positive
r =  0.0  → no relationship
r = -0.5  → moderate negative
r = -1.0  → perfect negative

Interpretation guide:
|r| >= 0.8  → very strong
|r| >= 0.6  → strong
|r| >= 0.4  → moderate
|r| >= 0.2  → weak
|r| <  0.2  → negligible

1. example
exam_score = 7.5 * study_hours + rng.normal(0, 7, n)
signal strength  = 7.5 × (range of study hours)
                 = 7.5 × (10-1) = 67.5 marks range (75-7.5)
noise strength   = std 7 marks ( noise only disturbs them by ±7 marks. )
signal >> noise → strong relationship → r will be high positive (~0.9)

2. example
sleep_hours = -0.3 * study_hours + rng.normal(7, 1, n)
signal strength = 0.3 × (range of study hours)
                = 0.3 × (10-1) = 2.7 hours range
noise strength  = std 1 hour
signal is only ~3× bigger than noise → weak relationship → r will be small negative (~-0.3)

**General Rule : 
large coefficient + small noise → strong correlation
small coefficient + large noise → weak correlation
**



````python

rng= np.random.default_rng(seed=42)
n=100
study_hrs= rng.uniform(1,10,n)#every value b/w 1 and 10 has equal chance of appearing
exam_score= 7.5 * study_hrs+ rng.normal(0,7,n)
exam_score=np.clip(exam_score,0,100)

sleep_hrs= -0.3 * study_hrs + rng.normal(7,1,n)
sleep_hrs= np.clip(sleep_hrs,3,10)
random= rng.normal(50,10,n)


def manual_cov(x,y,ddof=1):
    n=len(x)
    x_mean= np.mean(x)
    y_mean= np.mean(y)
    x_dev= x-x_mean
    y_dev= y-y_mean
    return np.sum(x_dev*y_dev)/(n-ddof)

cov_study_exam= manual_cov(study_hrs,exam_score)
cov_study_sleep= manual_cov(study_hrs,sleep_hrs)
cov_study_random= manual_cov(study_hrs,random)

print("Covariances:")
print(f"  study vs exam:   {cov_study_exam:.4f}  (positive → study more, score higher)")
print(f"  study vs sleep:  {cov_study_sleep:.4f} (negative → study more, sleep less)")
print(f"  study vs random: {cov_study_random:.4f}  (near zero → no relationship)")
#the numbers are in different scales so direct comparison is hard. This is exactly why we need correlation next.

#pearson correlation

def pearson_r(x,y):
    cov_xy= manual_cov(x,y)
    std_x= np.std(x,ddof=1)
    std_y=np.std(y,ddof=1)
    return cov_xy/((std_x)*(std_y))

r_study_exam= pearson_r(study_hrs,exam_score)
r_study_sleep= pearson_r(study_hrs,sleep_hrs)
r_study_random=pearson_r(study_hrs,random)

print("Pearson correlations: ")
print(f"Study vs Exam: {r_study_exam:.4f}")
print(f"Study vs Sleep: {r_study_sleep:.4f}")
print(f"Study vs Random: {r_study_random:.4f}")
print(" ")
for name , r in [("Study vs Exam", r_study_exam),("Study vs Sleep",r_study_sleep),
                 ("Study vs Random",r_study_random)]:
    if   np.abs(r)>=0.8: Strength="Very strong"
    elif np.abs(r)>=0.6: Strength= "Strong"
    elif np.abs(r)>=0.4: Strength= "Moderate"
    elif np.abs(r)>=0.2: Strength|= "Weak"
    else:                Strength="Negligible"
    direction= "Positive" if r>0 else  "Negative"
    print(f"{name}: {Strength}  {direction} correlation")  
    
## ── Full correlation matrix using np.corrcoef ────────────────

data_matrix= np.column_stack((study_hrs,exam_score,sleep_hrs,random))#(100,4)
correl_matrix= np.corrcoef(data_matrix.T)#(4,4)
names=["Study Hrs", "Exam Score", "Sleep Hrs", "Random"]
print(f"\n{' ':>12}",end="")
for name in names:
    print(f"{name:>12}",end="")
print()
for i , row_name in enumerate(names):
    print(f"{row_name:<12}",end="")
    for j in range(len(names)):
        val= correl_matrix[i,j]
        marker= "⭐" if np.abs(val)>=0.5 and i!=j else " "
        print(f"{val:>10.2f} {marker}",end="")
    print()
    
# ── Covariance matrix using np.cov ───────────────────────────
cov_matrix=np.cov(data_matrix.T)#(4,4)
print(f"\nCo variance matrix: Diagonal(variances):")
print(f"{cov_matrix.round(2)}")
print(f"Diagonal variances: {np.diag(cov_matrix).round(2)}")

# Verify: correlation = cov / (std_i * std_j)
std_vector= np.std(data_matrix,axis=0,ddof=1)
std_outer= np.outer(std_vector,std_vector)
corr_from_cov= cov_matrix/std_outer
print("Correlation from cov matrix:")
print(f"{corr_from_cov.round(2)}")
print(f"Match correl  matrixes: {np.allclose(correl_matrix,corr_from_cov)}")

# ── Finding strongest correlations automatically ─────────────
def find_strong_correlations(correl_matrix,names,threshold=0.3):
    correl_matrix=correl_matrix.round(2)
    n=len(names)
    strong_pairs=[]
    for i in range(n):
        for j in range(i+1,n):
            r=correl_matrix[i,j]
            if np.abs(r)>threshold:
                strong_pairs.append((names[i],names[j],r))
    strong_pairs.sort(key=lambda x: np.abs(x[2]),reverse=True)#Descending order sorting
    return strong_pairs
strong_pairs= find_strong_correlations(correl_matrix,names)
print(strong_pairs)

print("\nSignificant correlations (|r| > 0.2):")
print(f"{'Feature A':<14} {'Feature B':<14} {'r':>8}  {'Strength'}")
print("─" * 50)
for pairs in strong_pairs:
    r=np.abs(pairs[2])
    strength= "⭐⭐⭐" if r>0.8 else("⭐⭐" if r>0.6 else "⭐") 
    print(f"{pairs[0]:<14} {pairs[1]:<14} {pairs[2]:>8} {strength}")
    
                
    



    

````

**Density**

Raw counts depend on sample size:
class A: 200 students → bin 60-70 has count=45
class B:  50 students → bin 60-70 has count=11
bar heights are completely different
even if both classes have same score distribution
→ cannot compare directly

### What density does

Converts raw counts into a value where **total area = 1.0**
Formula:
**density = count / (total_count × bin_width)**
Example:
bin 60-70, count=45, total=200, bin_width=10
density = 45 / (200 × 10) = 0.0225

### Why area = 1.0 matters
**area of one bar = density × bin_width**
= 0.0225 × 10
= 0.225
meaning: 22.5% of students scored between 60-70
total area = sum of all bars = 1.0 = 100%
→ behaves like a probability distribution

### Verifying area = 1.0

```python
density, edges = np.histogram(data, bins=20, density=True)

# np.diff(edges) gives width of each bin
bin_widths = np.diff(edges)   # e.g. [5.0, 5.0, 5.0 ...]

# area = density × width for each bin
total_area = np.sum(density * bin_widths)
print(total_area)   # always exactly 1.0
```

Bootstrapping + Permutation Tests

## Part 1 — Bootstrapping

### The problem it solves
You have one sample of 50 scores and compute mean = 74.3.
How reliable is this estimate? Would a different 50 students give a very different mean?
Bootstrapping answers this using only the data you already have.

---

### Core idea — resampling with replacement
Pick n values randomly from your data, putting each back before the next pick.
Same value can appear multiple times. Same size as original.
original = [65, 72, 81, 58, 90]   n=5
resample 1 = [72, 72, 58, 90, 65]  → mean=71.4  (72 repeated)
resample 2 = [81, 90, 90, 74, 58]  → mean=78.6  (90 repeated)
resample 3 = [58, 65, 72, 81, 83]  → mean=71.8  (different mix)
...repeat 10,000 times...

**Why with replacement?**
Without replacement → same values every time → same mean every time → useless
With replacement → different values each time → different means → useful distribution

---

### Steps
```python
1. resample WITH replacement, same size n → 10,000 times
2. compute statistic (mean/median/anything) on each resample
3. sort the 10,000 statistics
4. cut 2.5% from each tail → 95% confidence interval
```

```python

bootstrap_samples = rng.choice(data, size=(10000, n), replace=True)
bootstrap_stats   = np.apply_along_axis(np.mean, axis=1, arr=bootstrap_samples)
ci_lower = np.percentile(bootstrap_stats, 2.5)
ci_upper = np.percentile(bootstrap_stats, 97.5)**
```

---

### Confidence interval
sort 10,000 bootstrap means:
[68.1 ... 71.3 ......................... 77.1 ... 79.4]
↑                               ↑
2.5th percentile            97.5th percentile
= ci_lower                  = ci_upper
95% CI = [71.3, 77.1]

---

### What confidence level means

| What people think | What it actually means |
|---|---|
| 95% chance true mean is in interval | If 100 researchers each built a CI, 95 would contain the true mean |
| interval moves | interval is fixed once computed — the TRUE MEAN is fixed |
true mean = 72.0 (fixed, unknown)
researcher 1: CI = [69.1, 78.3]  ✓ contains 72
researcher 2: CI = [70.2, 79.8]  ✓ contains 72
researcher 3: CI = [73.5, 82.1]  ✗ missed (bad luck)
...95 out of 100 intervals catch it

**Think of it as a fishing net:**
You cast a net. 95% confidence means your net catches the true value 95% of the time.

---

### Choosing confidence level
90% CI → alpha=5.0 → cut at 5.0 and 95.0   → narrower, less sure
95% CI → alpha=2.5 → cut at 2.5 and 97.5   → standard choice
99% CI → alpha=0.5 → cut at 0.5 and 99.5   → wider, more sure
**formula: alpha = (100 - confidence) / 2**

Higher confidence = cut less from tails = wider interval = more certain of catching true value.

---

### Works for ANY statistic — that is the power
analytical CI for median     → complex formula
analytical CI for correlation → even more complex
bootstrap CI for both         → exact same 5 lines of code
bootstrap_ci(sample, np.mean)    → CI for mean
bootstrap_ci(sample, np.median)  → CI for median
bootstrap_ci(sample, np.max)     → CI for maximum

---

### Effect of sample size
n=10  → CI width=12.4  → very uncertain
n=50  → CI width= 5.8  → more certain
n=200 → CI width= 2.9  → quite certain
n=500 → CI width= 1.8  → very certain
more data → narrower CI → more reliable estimate
theoretical: std_error = std / sqrt(n)
bootstrap agrees with this formula ← validates the approach

# Replace=True
rng.choice picks random values from data. The replace=True argument tells it to put each value back before picking the next one.
data = [65, 72, 81, 58, 90]

replace=False (without replacement):
pick 65 → remove it → [72, 81, 58, 90]
pick 72 → remove it → [81, 58, 90]
pick 81 → remove it → [58, 90]
pick 58 → remove it → [90]
pick 90 → done
result = [65, 72, 81, 58, 90] ← just a shuffle, same values always

replace=True (with replacement):
pick 72 → put it back → [65, 72, 81, 58, 90]
pick 72 → put it back → [65, 72, 81, 58, 90]  ← can pick 72 again
pick 58 → put it back → [65, 72, 81, 58, 90]
pick 90 → put it back → [65, 72, 81, 58, 90]
pick 65 → put it back → done
result = [72, 72, 58, 90, 65] ← 72 repeated, 81 missing
Without replace=True every resample would just be a shuffle of the original — same mean every time — completely useless for bootstrapping.
replace=True is what creates the variation between resamples that makes bootstrapping work.



````python
rng= np.random.default_rng(seed=42)
sample= rng.normal(loc=70,scale=15,size=50)
np.clip(sample,0,100)
print(f"Sample Size:{len(sample)}")
print(f"Sample mean:{np.mean(sample):.4f}")
print(f"Sample std:{np.std(sample,ddof=1):.4f}")

# ── Bootstrap implementation ─────────────────────────────────
def bootstrap_ci(data,statistic_fn,n_bootstrap=10000,confidence=95):
    bootstrap_samples= rng.choice(data,size=(n_bootstrap,len(data)),replace=True)#(10,000,50)
    bootstrap_stats= np.apply_along_axis(statistic_fn,axis=1,arr= bootstrap_samples)
    alpha= (100-confidence)/2 # we cut alpha from head and tail
    ci_lower= np.percentile(bootstrap_stats,alpha)
    ci_upper= np.percentile(bootstrap_stats,100-alpha)
    
    
    return {
        'point_estimate':          round(float(statistic_fn(data)), 6),
        'ci_lower':                round(float(ci_lower), 6), # converts number from numpy scalar to python float which is feasible for round function
        'ci_upper':                round(float(ci_upper), 6),
        'bootstrap_std':           round(float(np.std(bootstrap_stats)), 6),
        'bootstrap_distribution':  bootstrap_stats,
        'confidence':              confidence
    }
# ── Bootstrap CI for multiple statistics ────────────────────
statistics= {
             "mean":np.mean,
             "median":np.median,
             "Std Dev": lambda x: np.std(x,ddof=1),
             "Q1(25%)": lambda x: np.percentile(x,25),
             "Q3 (75%)": lambda x:np.percentile(x,75),
             "max"     : np.max,
             
}
print("Bootstrap 95% Confidence Intervals")
print(f"{'Statistic':<14} {'Estimate':>10} {'CI Lower':>10} "
      f"{'CI Upper':>10} {'Width':>8}")
print("─" * 56)

for stat_name,stat_fn in statistics.items():
    result= bootstrap_ci(sample,stat_fn)
    width=result["ci_upper"] -result["ci_lower"]
    print(f"{stat_name:<14} {result['point_estimate']:>10} {result['ci_lower']:>10}",
          f"{result['ci_upper']} {width:>8.2f}")  
````

Standard error:
It measures how spread out the sample means are from the true population mean.

**Formula : std_error = std / sqrt(n)**

population: all students in India (millions)
true mean exam score = 72.0

you collect sample 1 (n=50) → mean = 71.3
you collect sample 2 (n=50) → mean = 73.1
you collect sample 3 (n=50) → mean = 70.8
you collect sample 4 (n=50) → mean = 72.5
...

these sample means vary around 72.0
how much they vary = standard error

1. CI width nearly halves when n quadruples
n=10  → 22.74
n=50  → 7.33   (5× more data → width dropped by 3×)
n=200 → 4.56   (4× more data → width dropped by ~2×)
n=500 → 2.56

This matches the formula std_error = std/sqrt(n) — to halve the error you need 4× more data (because sqrt(4)=2).

2. Bootstrap CI width ≈ 4 × std_error
n=50:  CI width = 7.33,  std_error = 1.91,  ratio = 7.33/1.91 = 3.84 ≈ 4 ✓
n=100: CI width = 6.08,  std_error = 1.54,  ratio = 6.08/1.54 = 3.95 ≈ 4 ✓
n=200: CI width = 4.56,  std_error = 1.15,  ratio = 4.56/1.15 = 3.97 ≈ 4 ✓
Bootstrap agrees with mathematical theory ✓ — this validates the bootstrap approach.

````python
# ── Effect of sample size on CI width ───────────────────────
# The larger the sample, the narrower the CI (more certainty)
print("Effect of sample size on CI width (for mean)")
print(f"{'Sample_size':>12} {'CI Width':>10} {'Std Error':>12}")
print("-"*36)

for n_size in [10, 25, 50, 100, 200, 500]:
    sub_sample= rng.normal(loc=70,scale=15,size=n_size)
    result=bootstrap_ci(sub_sample,np.mean,n_bootstrap=5000)
    width= result["ci_upper"]-result["ci_lower"]
    std_error= np.std(sub_sample,ddof=1)/np.sqrt(n_size)
    print(f"{n_size:>12} {width:>10.2f} {std_error:>12.2f}")
    
````
